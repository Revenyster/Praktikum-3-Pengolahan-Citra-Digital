```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

Task 1

```python
rainbow=cv2.imread(r"D:\Photos\pcd312.jpg") 
showr=cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB) 
```

```python
img=rainbow 
img.shape 
mask=np.zeros(img.shape[:2],np.uint8) 
mask[300:400,100:400]=255 
plt.imshow(mask,cmap='gray') 
```

```python
plt.imshow(img) 
```

```python
plt.imshow(showr) 
```

```python
masked_img=cv2.bitwise_and(img,img,mask=mask) 
show_masked_img=cv2.bitwise_and(showr,showr,mask=mask) 
plt.imshow(show_masked_img)
```

Task 2

```python
hist_mask_value_red=cv2.calcHist([rainbow],channels=[2],mask=mask,histSize=[256],ranges=[0,256]) 
hist_value_red=cv2.calcHist([rainbow],channels=[2],mask=None,histSize=[256],ranges=[0,256]) 
```

```python
plt.plot(hist_mask_value_red) 
plt.title('red histogram for masked rainbow') 
```

```python
plt.plot(hist_value_red) 
plt.title('red histogram for normal rainbow')
```
